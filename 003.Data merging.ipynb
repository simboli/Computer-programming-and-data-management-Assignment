# Modules

```python
import pandas as pd
```

# Variables

```python
ds_EU27_emissions = '02.Transformed data/01.EU27_emissions.csv'
ds_EU27_passengers = '02.Transformed data/02.EU27_passengers.csv'
```

```python
out_EU27_emissions_proCapita = '02.Transformed data/03.EU27_emissions_proCapita.csv'
```

# Data load

```python
df_EU27_emissions=pd.read_csv(ds_EU27_emissions,header=0, dtype='unicode')
```

```python
cols = df_EU27_emissions.columns
for col in cols:
    if col != 'Year':
        df_EU27_emissions[col] = df_EU27_emissions[col].astype(float)

df_EU27_emissions['Year'] = df_EU27_emissions['Year'].astype("int")
```

```python
df_EU27_passengers=pd.read_csv(ds_EU27_passengers,header=0, dtype='unicode')
```

```python
df_EU27_passengers['Year'] = df_EU27_passengers['Year'].astype("int")
df_EU27_passengers['Passengers'] = df_EU27_passengers['Passengers'].astype("int")
```

```python
df_EU27_emissions.head()
```

```python
df_EU27_passengers.head()
```

# Data merging and pro passenger emissions calculation

```python
df_EU27_emissions_proCapita = pd.merge(df_EU27_emissions, df_EU27_passengers, on='Year')
```

```python
df_EU27_emissions_proCapita.head()
```

```python
#df_EU27_emissions_proCapita['As'] = df_EU27_emissions_proCapita['As']/df_EU27_emissions_proCapita['Passengers']
```

```python
# Iterate over column names
for column in df_EU27_emissions_proCapita:
    if column != 'Year' and column != 'Passengers':
        df_EU27_emissions_proCapita[column] = df_EU27_emissions_proCapita[column]/df_EU27_emissions_proCapita['Passengers']
```

```python
df_EU27_emissions_proCapita.head()
```

```python
df_EU27_emissions_proCapita = df_EU27_emissions_proCapita.drop('Passengers', axis=1)
```

# Exporting

```python
df_EU27_emissions_proCapita.to_csv(out_EU27_emissions_proCapita, index=False)
```
